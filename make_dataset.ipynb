```python
path = '/content/drive/MyDrive/project/package'
import sys
sys.path.append(path)  
from gluonts.dataset.repository.datasets import get_dataset, dataset_recipes
from gluonts.dataset.util import to_pandas
import pandas as pd
import os
```

```python
print(f"Available datasets: {list(dataset_recipes.keys())}")
```

```python
def get_csv(data_name, n_features):
    dataset = get_dataset(data_name + '_nips', regenerate=False)
    # n_features = int(dataset.metadata['cardinality'])
    print(dataset.metadata)
    list_df = []
    temp_list = []
    for i, data in enumerate(dataset.test):
        df = to_pandas(data)
        temp_list.append(df)
        if (i+1)%n_features == 0:
            df_temp = pd.concat(temp_list, axis=1)
            list_df.append(df_temp)
            temp_list = []
    target_df = list_df[-1]

    assert(len(target_df.columns)==n_features)
    foldername = '/content/drive/MyDrive/project/CSDI_forecast/dataset/csv/'
    os.makedirs(foldername, exist_ok=True)
    target_df.to_csv(foldername + data_name +'.csv')
```

```python
list_data = ['solar', 'electricity', 'traffic']
list_n_features = [137,370,963]
for data, n_features in zip(list_data, list_n_features):
    get_csv(data, n_features)
```

```python
def get_csv_taxi(n_features=1214):
    dataset = get_dataset('taxi_30min', regenerate=False)
    print(dataset.metadata)
    #test data
    list_df = []
    temp_list = []
    for i, data in enumerate(dataset.test):
        df = to_pandas(data)
        temp_list.append(df)
        if (i+1)%n_features == 0:
            df_temp = pd.concat(temp_list, axis=1)
            list_df.append(df_temp)
            temp_list = []
    test_df = list_df[-1]
    test_df = test_df[:48+24*56]
    
    assert(len(test_df.columns)==n_features)
    foldername = '/content/drive/MyDrive/project/CSDI_forecast/dataset/csv/'
    os.makedirs(foldername, exist_ok=True)
    test_df.to_csv(foldername +'taxi_test.csv')

    #training data
    list_df = []
    for data in dataset.train:
        df = to_pandas(data)
        list_df.append(df)        
    train_df = pd.concat(list_df, axis=1)

    assert(len(train_df.columns)==n_features)
    train_df.to_csv(foldername +'taxi_train.csv')
    
```

```python
def get_csv_wiki(n_features=9535):
    dataset = get_dataset('wiki-rolling_nips', regenerate=False)
    print(dataset.metadata)
    list_df = []
    temp_list = []
    for i, data in enumerate(dataset.test):
        df = to_pandas(data)
        temp_list.append(df)
        if (i+1)%n_features == 0:
            df_temp = pd.concat(temp_list, axis=1)
            list_df.append(df_temp)
            temp_list = []
    target_df = list_df[-1]
    target_df = target_df.iloc[:,:2000]

    assert(len(target_df.columns)==2000)
    foldername = '/content/drive/MyDrive/project/CSDI_forecast/dataset/csv/'
    os.makedirs(foldername, exist_ok=True)
    target_df.to_csv(foldername + 'wiki.csv')
```
